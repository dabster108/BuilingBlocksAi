```python
df = pd.read_csv('/content/sample_data/btc-market-price.csv',
                 header=None,
                 na_values=['', '?', '-'],
                 names=['Timestamp', 'Price'],
                 dtype= {'Price' : 'float'})
df.head()
print(df.dtypes)
```

```python
#Date parser using parse_dates paramete
df = pd.read_csv('/content/sample_data/btc-market-price.csv',
                 header=None,
                 na_values=['', '?', '-'],
                 names=['Timestamp', 'Price'],
                 dtype={'Price': 'float'},
                 parse_dates=[0])
df.head()
```
