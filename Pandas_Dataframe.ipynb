<h1><b> # Dropna Parameters

| Parameter | Description |
|-----------|-------------|
| `how='any'` | Drops a row/column if any value is `NaN`. |
| `how='all'` | Drops a row/column only if all values are `NaN`. |
| `thresh=N` | Keeps rows/columns with at least `N` non-null values. |
| `axis=0` (default) | Drops rows. |
| `axis=1` | Drops columns. |
| `subset=['col1', 'col2']` | Drops rows only if `NaN` appears in specified columns. |
